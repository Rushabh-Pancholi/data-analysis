## Defining functions

```python
def beatify(ax,data,right_hide=True):
    
    ax.spines['bottom'].set_color('#CCCCCC')
    ax.spines['left'].set_color('#CCCCCC')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_color('#CCCCCC')
    ax.spines['right'].set_visible(False)
   
    if right_hide==False:
        ax.spines['right'].set_visible(True)
   
    ax.grid(False)
    ncount = len(data)
   
    ax.tick_params(axis="x", colors='#3A3A3B')
    ax.tick_params(axis="y", colors='#3A3A3B')
    
    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), ha='center', va='bottom',fontsize=11)

def counter(feature,ax=None,color=None,data=terr_data,orient='h'):

    if color==None:
        ax = sns.countplot(x= feature,data=data,ax=ax)
    else:
        ax = sns.countplot(x= feature,data=data,ax=ax,palette=color,order=data[feature].value_counts().index)

    beatify(ax,data)
        
    ax.set_title(data[feature].describe().name, size=14, pad=13)
    plt.tight_layout()
```

```python
def cntry_wise(top_by,ax=None,colors=None, country_wise = True):
    
    if country_wise == True:
        country_data = terr_data.groupby('Country',as_index=False).sum()
        data =  country_data.sort_values(by=top_by,ascending=False)[:20]
        sns.barplot(x='Country',y=top_by,data=data,palette=colors)
    else:
        country_data = terr_data.groupby('City',as_index=False).sum()
        data =  country_data.sort_values(by=top_by,ascending=False)[:20]
        sns.barplot(x='City',y=top_by,data=data,palette=colors)
    
    plt.xticks(rotation=90)
    plt.title('Top 20 Attacked Country In The World')
    beatify(ax,terr_data)
```

```python
def country(country_name,color=None,pie_color=None,linecolor=None):

## Data Preparation
    
    c_data =  terr_data[terr_data['Country']==country_name]
    region = c_data['Region'].unique()
    
    year_data = c_data.groupby(by='Year',as_index=False).sum()
    ci_data = c_data.groupby(by='City',as_index=False).sum().sort_values(by='Attack',ascending=False)
    
    City_data = c_data.groupby(by='City',as_index=False).sum().sort_values(by='Attack',ascending=False)[:10]
    
    region_count = terr_data[terr_data['Region']==region[0]]
    
    per = len(c_data)/len(terr_data)


## 1st Plot :-

    fig,axs = plt.subplots(ncols=2,figsize=(18,3))
    
    sns.lineplot(data=year_data,x='Year',y='Attack',ax=axs[0],color=linecolor)
    axs[0].set_title('Yearly Attacks In '+country_name)

    sns.barplot(x='City',y='Attack',data=City_data, ax=axs[1],palette=color)
    beatify(axs[1],c_data)
    axs[1].set_title('Top 10 Most Attacked Cities',fontsize=14)
    axs[1].set_xticklabels(axs[1].get_xticklabels(),rotation=60,size=12)


## 2nd Plot :-

    fig,axs = plt.subplots(ncols=2,figsize= (16,5), gridspec_kw=dict(width_ratios=[5,2]))
    
    axs[0].set_xticklabels(axs[0].get_xticklabels(),rotation=90,size=11)
    counter('Target Type',data= c_data,ax=axs[0],color=color)
    
    plt.pie((per,1-per),labels=[country_name,'Other'],colors=pie_color,autopct='%1.1f%%',textprops={'fontsize':12},wedgeprops={'linewidth': 1, 'linestyle': 'solid', 'antialiased': True,'edgecolor':'Black'},radius=1.2)
    plt.title('Percentage Attack of whole Region',fontsize=14)
    
    plt.tight_layout()


## 3rd Plot :

    fig,axs = plt.subplots(ncols=2,figsize=(16,5),gridspec_kw=dict(width_ratios=[3,5]))
    
    sns.countplot(y='Attack Type',data=c_data,ax = axs[0],order=c_data['Attack Type'].value_counts().index,palette=color)
    axs[0].set_title('Type of attack',fontsize=14)
    
    counter('Weapon Type',data=c_data,ax = axs[1],color=color)
    axs[1].set_xticklabels(axs[1].get_xticklabels(),rotation=60,size=11)
    
    plt.tight_layout()
```
